```python
dfCombos = pd.read_csv('proc_mod_diag_combos.csv')
dfCombos['proc_values'] = dfCombos['proc_values'].astype(str)
dfCombos['proc_sorted'] = dfCombos['proc_values'].str.split('@')
dfCombos['proc_sorted'] = dfCombos['proc_sorted'].apply(sorted)
dfCombos['proc_sorted'] = dfCombos['proc_sorted'].transform(lambda x: '@'.join(x))
dfCombos['proc_values'] = dfCombos['proc_sorted']

dfCombos['diag_values'] = dfCombos['diag_values'].astype(str)
dfCombos['diag_sorted'] = dfCombos['diag_values'].str.split('@')
dfCombos['diag_sorted'] = dfCombos['diag_sorted'].apply(sorted)
dfCombos['diag_sorted'] = dfCombos['diag_sorted'].transform(lambda x: '@'.join(x))
dfCombos['diag_values'] = dfCombos['diag_sorted']
```

```python
dfCombos['mod_values'] = dfCombos['mod_values'].astype(str)
dfCombos = dfCombos[dfCombos['proc_values'].str.contains('97032')]
dfCombos = dfCombos[['claim_id', 'proc_values',
                     'diag_values', 'mod_values']].groupby(
    ['claim_id', 'proc_values', 'diag_values'])
dfCombos = dfCombos['mod_values'].apply(lambda x: '@'.join(x)).reset_index()
```

```python
dfCombos['mod_sorted'] = dfCombos['mod_values'].str.split('@')
dfCombos['mod_sorted'] = dfCombos['mod_sorted'].apply(sorted)
dfCombos['mod_sorted'] = dfCombos['mod_sorted'].transform(lambda x: '@'.join(x))
dfCombos['mod_values'] = dfCombos['mod_sorted']
dfCombos = dfCombos.drop(['mod_sorted'], axis=1)
```
