```python
import numpy as np

import pandas as pd

```

```python
df1 = pd.read_csv("chiro_10.csv")

df1.head()
```

```python
df = pd.DataFrame()
dg = []

for cpt in df1.columns:
    print('working on cpt: ' + str(cpt))
    df_sample = df1[[cpt]].copy()

    d = pd.DataFrame()
    i = 0
    while i < 10:
        dfa = df_sample.sample(n=3000)
        a = dfa.mean(axis=0)
        a = np.round(a, 2)
        t = pd.DataFrame(
            {
                'adv': a
            }
        )
    
        i+=1
        d = pd.concat([d, t])
    a = d.mean()
    a = float(a)
    b = d.std()
    b = float(b)

    data = pd.read_csv("s_chiro_10.csv")
    mu, sigma = a , b

    s = np.random.normal(mu, sigma, len(data))
    dg = pd.DataFrame(s, columns={cpt})
    df[cpt] = dg[cpt]
df.head()
```
