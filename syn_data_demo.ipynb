```python
import pandas as pd
pd.set_option('display.max_rows', 1000)
```

```python
df = pd.read_csv('proc_diag_combos_chiro (1).csv')
df = df.fillna('None')
```

```python
df.head()
```

```python
#### Find percentage prevalence for each combination within a specific code
def calculate_percentage(df):
    total_lines = df['line_cnt'].sum()
    total_providers = df['prov_cnt'].sum()
    df['pct_composition'] = df['line_cnt']/total_lines
    df['pct_peers'] = df['prov_cnt']/total_providers
    return df
```

```python
df.groupby('hcpcs_code').apply(calculate_percentage)
```
